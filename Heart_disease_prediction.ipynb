```python
x = heart_data.drop(columns = 'target', axis=1)
y = heart_data['target']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
import joblib
joblib.dump(scaler, 'scaler.pkl')
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
import pickle
import joblib

with open('heart_disease_model.pkl', 'wb') as f:
  pickle.dump(model,f)
  print("model saved as heart_disease_model.pkl using pickle")

joblib.dump(model,'heart_disease_model.joblib')
print("Model saved as heart_disease_model.joblib using joblib")
```

```python
import plotly.graph_objects as go

def generate_gauge(risk_probability):
    fig = go.Figure(go.Indicator(
        mode="gauge+number+delta",
        value=risk_probability,
        number={'valueformat': '.2%'},
        domain={'x': [0, 1], 'y': [0, 1]},
        title={'text': " Risk Probability", 'font': {'size': 24}},
        delta={'reference': 0.5, 'increasing': {'color': "red"}, 'decreasing': {'color': "green"}},
        gauge={
            'axis': {'range': [0, 1], 'tickformat': '.0%'},
            'bar': {'color': "indigo"},
            'steps': [
                {'range': [0, 0.25], 'color': "lightgreen"},
                {'range': [0.25, 0.5], 'color': "gold"},
                {'range': [0.5, 0.75], 'color': "orange"},
                {'range': [0.75, 1], 'color': "red"}
            ],
            'threshold': {
                'line': {'color': "black", 'width': 4},
                'thickness': 0.75,
                'value': risk_probability
            }
        }
    ))
    fig.update_layout(height=400, margin=dict(t=50, b=0, l=50, r=50))
    return fig
```

```python
import plotly.graph_objects as go
def predict_heart_disease_with_chart(age, sex, cp, trestbps, chol, fbs, restecg,
                                     thalach, exang, oldpeak, slope, ca, thal):
    input_data = np.array([age, sex, cp, trestbps, chol, fbs, restecg,
                           thalach, exang, oldpeak, slope, ca, thal])
    input_data = input_data.reshape(1, -1)

    scaler = joblib.load('scaler.pkl')
    input_data_scaled = scaler.transform(input_data)

    prob = model.predict_proba(input_data_scaled)[0][1]

    fig = go.Figure(go.Indicator(
        mode = "gauge+number+delta",
        value = prob,
        title = {'text': " Heart Disease Risk Meter", 'font': {'size': 24, 'color': 'indigo'}},
        delta = {'reference': 0.5},
        gauge = {
            'axis': {'range': [0, 1]},
            'bar': {'color': "darkviolet"},
            'steps': [
                {'range': [0, 0.4], 'color': '#c2f0c2'},
                {'range': [0.4, 0.7], 'color': '#fff4b3'},
                {'range': [0.7, 1.0], 'color': '#ffcccc'}
            ],
            'threshold': {
                'line': {'color': "red", 'width': 6},
                'thickness': 0.8,
                'value': prob
            }
        }
    ))
    fig.update_layout(
    paper_bgcolor='#fdf6e3',  # outer background (beige)
    plot_bgcolor='#fdf6e3',   # inner plotting area background (same beige)
)

    result = " At Risk" if prob >= 0.4 else " Healthy"
    chart = generate_gauge(prob)

    return f"{result} | Probability: {prob:.2%}", chart
```

```python
import gradio as gr
custom_css = """
body {
    font-family: 'Poppins', sans-serif;
    background-color: #ffffff;
    color: #3b3c7d;
}

label {
    font-weight: bold !important;
    color: #3b3c7d !important;
}

input[type="number"], select, textarea {
    background-color: 	#3b3c7d !important;
    color: white !important;
    border: 1px solid #3b3c7d !important;
    border-radius: 5px;
    padding: 5px;
}

.gr-button {
    background-color: #3b3c7d !important;
    color: white !important;
    font-weight: bold;
    border-radius: 5px;
}
"""

with gr.Blocks(css=custom_css) as demo:
    gr.Markdown("##  Heart Disease Risk Predictor")

    with gr.Row():
        with gr.Column():
            age = gr.Number(label="Age")
            sex = gr.Dropdown([0, 1], label="Sex (0=Female, 1=Male)")
            cp = gr.Dropdown([0,1,2,3], label="Chest Pain Type")
            trestbps = gr.Number(label="Resting BP")
            chol = gr.Number(label="Cholesterol")
            fbs = gr.Dropdown([0,1], label="Fasting Blood Sugar > 120mg/dl")
            restecg = gr.Dropdown([0,1,2], label="Resting ECG Result")
        with gr.Column():
            thalach = gr.Number(label="Max Heart Rate")
            exang = gr.Dropdown([0,1], label="Exercise Induced Angina")
            oldpeak = gr.Number(label="ST Depression")
            slope = gr.Dropdown([0,1,2], label="Slope of ST")
            ca = gr.Dropdown([0,1,2,3,4], label="Number of Major Vessels")
            thal = gr.Dropdown([0,1,2,3], label="Thalassemia")

    predict_btn = gr.Button("Predict ")
    output = gr.Textbox(label="Prediction", interactive=False)
    chart_output = gr.Plot(label=" Risk Gauge")

    predict_btn.click(
        fn=predict_heart_disease_with_chart,
        inputs=[age, sex, cp, trestbps, chol, fbs, restecg,
                thalach, exang, oldpeak, slope, ca, thal],
        outputs=[output,chart_output]
    )

demo.launch()
```
